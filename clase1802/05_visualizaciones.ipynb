# Visualizaciones

## Matplotlib

Matplotlib es una de las bibliotecas más populares en Python para la creación de gráficos y visualizaciones de datos. Fundada en 2003 por John D. Hunter, proporciona herramientas robustas para generar gráficos bidimensionales de alta calidad. Es fundamental para cualquiera que trabaje con datos en Python, siendo la base sobre la cual se han construido otras bibliotecas como Seaborn y Plotly.

Matplotlib se destaca por su simplicidad y flexibilidad. Aunque existen otras bibliotecas para la visualización de datos, Matplotlib sigue siendo la más utilizada gracias a su capacidad para generar gráficos altamente personalizables. Además, al ser de código abierto, es accesible y ampliamente utilizada por la comunidad científica y de desarrollo.

```python
data = {
    "fecha_compra": pd.date_range(start="2024-01-01", periods=12, freq="15D"),
    "mes": ["Ene", "Ene", "Ene", "Feb", "Feb", "Feb", "Mar", "Mar", "Mar", "Abr", "Abr", "Abr"],
    "sexo": ["H", "M", "H", "H", "M", "M", "H", "H", "M", "M", "H", "M"],
    "consumo": [50, 60, 70, 90, 110, 100, 120, 130, 150, 160, 140, 180],
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# Agrupar por mes y sexo sumando los consumos
df_grouped = df.groupby(["mes", "sexo"])["consumo"].sum().unstack()
df_grouped["total"] = df_grouped["H"] + df_grouped["M"]

df_grouped
```

```python
# Crear gráfico apilado
x = np.arange(len(df_grouped.index))  # Posiciones en el eje X

plt.figure(figsize=(8, 5))

plt.bar(x, df_grouped["H"], label="Hombres", color="blue")
plt.bar(x, df_grouped["M"], bottom=df_grouped["H"], label="Mujeres", color="green")  # Se apila sobre Hombres

# Configurar etiquetas y título
plt.xticks(x, df_grouped.index)
plt.xlabel("Mes")
plt.ylabel("Consumo Total")
plt.title("Consumo Mensual por Sexo (Apilado)")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```

Gtrafico de Tortas - Pie chart

Se utilizan para representar proporciones o porcentajes de un todo. Son útiles cuando se quieren mostrar cómo se distribuyen las partes de un conjunto total.

```python
# Datos de ejemplo -- Unidades de vehiculos vendidas por marca
data = {
    "Fabricante": ["Toyota", "Volkswagen", "Ford", "Honda", "BMW", "Chevrolet"],
    "Unidades": [25000, 20000, 15000, 13000, 12000, 15000]
}

# Crear DataFrame
df = pd.DataFrame(data)
df
```

```python
plt.pie(df["Unidades"], labels=df["Fabricante"], autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    df['Unidades'],
    labels=df["Fabricante"],
    autopct='%1.1f%%',   # Muestra los porcentajes con 1 decimal
    startangle=140,      # Ángulo inicial para mejorar la distribución
    wedgeprops={'edgecolor': 'black', 'linewidth': 1},  # Bordes negros para mejor visibilidad
    textprops={'fontsize': 10}  # Tamaño del texto
)

# Agregar título
plt.title("Participacion de mercado de marcas de autos", fontsize=14, fontweight="bold")

plt.show()
```

```python
# Datos simulados
data = {
    "Marca": ["Toyota", "Volkswagen", "Ford", "Honda", "BMW", "Chevrolet"],
    "Ventas": [25000, 20000, 15000, 13000, 12000, 15000]
}
df = pd.DataFrame(data)

# Ordenar de mayor a menor
df = df.sort_values(by="Ventas", ascending=False)

# Extraer datos ordenados
marcas = df["Marca"]
ventas = df["Ventas"]

# Graficar Pie Chart ordenado
plt.figure(figsize=(8, 8))
plt.pie(ventas, labels=marcas, autopct="%1.1f%%", startangle=180, colors=["red", "blue", "green", "purple", "orange", "gray"])

# Título
plt.title("Participación de Mercado de Marcas de Autos (Ordenado)")
plt.show()
```

Boxplot

Ayuda a visualizar la distribución de datos y detectar **outliers.**

Caja (box):

La caja se dibuja entre el primer cuartil (Q1) y el tercer cuartil (Q3). El primer cuartil es el valor que divide el 25% inferior de los datos, y el tercer cuartil divide el 25% superior. Es decir, el 50% de los datos están dentro de la caja. La longitud de la caja es la rango intercuartílico (IQR), que es la diferencia entre Q3 y Q1. Esto representa la dispersión de la mitad central de los datos. Línea en el medio de la caja (mediana):

La línea dentro de la caja es la mediana de los datos (también conocida como el segundo cuartil, Q2). La mediana divide a los datos en dos partes iguales, es el valor central cuando los datos están ordenados. Bigotes (whiskers):

Los bigotes se extienden desde los cuartiles hasta el valor máximo y el valor mínimo dentro de un rango aceptable. Este rango es generalmente 1.5 veces el rango intercuartílico (IQR). Cualquier valor fuera de este rango se considera un outlier (valor atípico). Outliers:

Son los puntos que se encuentran fuera de los bigotes (más allá de 1.5 veces el IQR). Estos valores pueden indicar errores de medición o variaciones inusuales en los datos.

```python
# Datos simulados: alturas de personas en cm
alturas = np.random.normal(loc=170, scale=10, size=1000)  # Media=170 cm, Desviación=10 cm
```

```python
alturas
```

```python


# Crear el Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(alturas, vert=False, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'),
            whiskerprops=dict(color='black', linewidth=1.5),
            flierprops=dict(marker='o', color='red', markersize=5))

# Añadir etiquetas y título
plt.xlabel('Altura (cm)')
plt.title('Boxplot de Alturas de Personas')

plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.show()
```
