```python
import pandas as pd

# Crear un DataFrame con varias filas y columnas de tipo numérico
data = {
    "columna_1": [1000.9, 2000, 3000, 4000, 5000],
    "columna_2": [10000, 20000, 30000, 40000, 50000],
    "columna_3": [10000000, 2000000.8, 30000, 400000, 500000]
}

df = pd.DataFrame(data)

# Aplicar formato con separador de miles a la columna
df['columna_3']= df['columna_3'].apply(lambda x: f"{x:,}")
df["columna_3"]
```

**Nota**: El separador de miles convierte la columna a tipo object(str) usar solo para mostrar los datos. Luego convertir nuevamente a enteros

Para graficos

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Crear un DataFrame de ejemplo
df = pd.DataFrame({'valor': [1000000, 2500000, 300000]})

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['valor'], marker='o')

# Formatear el eje Y con separadores de miles
formatter = FuncFormatter(lambda x, pos: f'{int(x):,}')
plt.gca().yaxis.set_major_formatter(formatter)

# Mostrar gráfico
plt.title('Gráfico con Separadores de Miles')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
```
