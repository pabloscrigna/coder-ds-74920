```python
import pandas as pd

# Crear un DataFrame con varias filas y columnas
data = {
    "columna_1": ["1", "2", "3", "4", "5"],
    "columna_2": ["10", "20", "30", "40", "50"],
    "columna_3": ["100", "200", "300", "400", "500"]
}

df = pd.DataFrame(data)
df
```

```python
# Cambiar el valor en la fila 2, columna "columna_2"
df.at[2, "columna_2"] = "999"
df
```

```python
# Otra forma usando loc
df.loc[2, "columna_3"] = "888"
```
