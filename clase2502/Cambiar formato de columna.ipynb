Convertir una columna object a int64

```python
import pandas as pd

# Creamos una lista de enteros de tipo str
data = ['1', '2', '3', '4', '5']

# Creamos un DF
df = pd.DataFrame({'notas': data})
```

```python
df
```

```python
df.dtypes
```

```python
# pasamos la columna a tipo int
df['notas'] = df['notas'].astype(int)
```
